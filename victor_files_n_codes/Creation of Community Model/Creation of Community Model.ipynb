```python
import re

Info_CSC009 = ""
Info_CSC031 = ""
Info_CSC040 = ""
Info_CSC043 = ""
Info_CSC052 = ""
for reaction in com.reactions:
    if re.search(".+\:(.+)\<\=\>\s\d\.\d\d\s(.+_m)", str(reaction)):
        match1 = re.search(".+\:(.+)\<\=\>\s\d\.\d\d\s(.+_m)", str(reaction))
        if "CSC031" in match1.group(1):
            Info_CSC031 = Info_CSC031 + match1.group(1) + "\t" + match1.group(2) + "\n"
        elif "CSC040" in match1.group(1):
            Info_CSC040 = Info_CSC040 + match1.group(1) + "\t" + match1.group(2) + "\n"
        elif "CSC043" in match1.group(1):
            Info_CSC043 = Info_CSC043 + match1.group(1) + "\t" + match1.group(2) + "\n"
        elif "CSC052" in match1.group(1):
            Info_CSC052 = Info_CSC052 + match1.group(1) + "\t" + match1.group(2) + "\n"
        elif "CSC009" in match1.group(1):
            Info_CSC009 = Info_CSC009 + match1.group(1) + "\t" + match1.group(2) + "\n"
new_file_CSC031 = open("Exchanged_metabolites_CSC031_NoCurtobacterium.tsv", "w")
new_file_CSC031.write(Info_CSC031)
new_file_CSC040 = open("Exchanged_metabolites_CSC040_NoCurtobacterium.tsv", "w")
new_file_CSC040.write(Info_CSC040)
new_file_CSC043 = open("Exchanged_metabolites_CSC043_NoCurtobacterium.tsv", "w")
new_file_CSC043.write(Info_CSC043)
new_file_CSC052 = open("Exchanged_metabolites_CSC052_NoCurtobacterium.tsv", "w")
new_file_CSC052.write(Info_CSC052)
#print(Info_CSC031)
#print(Info_CSC040)
#print(Info_CSC043)
#print(Info_CSC052)
#print(Info_CSC009)
```

```python
solution = com.optimize(fluxes=True, pfba=True)
fluxes_solution = solution.fluxes
columns = fluxes_solution.columns
column_name_list = columns.tolist()

new_column_list = []
for column_name in column_name_list:
    if "EX_" in column_name and "_m" in column_name:
        new_column_list.append(column_name)
#print(len(new_column_list))

filtered_fluxes_solution = fluxes_solution.filter(new_column_list)
#print(len(filtered_fluxes_solution.iloc[4].values))
#print(len(new_column_list))

nozeros_list = []
for i in range(len(filtered_fluxes_solution.iloc[4].values)):
    if filtered_fluxes_solution.iloc[4].values[i] != 0:
        nozeros_list.append(new_column_list[i])
#print(nozeros_list)
nozeros_fluxes_solution = filtered_fluxes_solution.filter(nozeros_list)

for i in range(len(nozeros_fluxes_solution.iloc[4].values)):
    print(str(nozeros_list[i]) + "\t" + str(nozeros_fluxes_solution.iloc[4].values[i]) + "\n")

#print(nozeros_fluxes_solution)
#print(filtered_fluxes_solution.iloc[0])

com_summary = com.summary()
#print(type(com_summary))

values = nozeros_fluxes_solution.iloc[4]
#print(values)
```
