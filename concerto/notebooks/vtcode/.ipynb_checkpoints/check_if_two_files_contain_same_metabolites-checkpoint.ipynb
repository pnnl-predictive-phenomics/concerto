# 2. This code will help you look for similar components in two different files.
Compares carbon sources of observed growth (Biolog compounds) vs. metabolites observed to be consumed/produced
a. which metabolites can organisms consume that are produced by other organism? -> different conditions in metabolomics data
b. which metabolites can organisms grow (can be used as a carbon source) on that are produced by other organisms? if an organism can consume this metabolite, can it also grow on this metabolite as a single carbon source (Biolog vs biolog data)
  *want to be able to reproduce these observations with our models

```python
#We import pandas and regex
import pandas as pd
import re

#We load the files that we want to compare, in this case it will be the biolog file and the exchange metabolomics file.
#Column header in Biolog file MUST have "Compound" and in the BiocycID file MUST contain "Metabolites"
biolog = pd.read_csv("C:\\Users\\lint730\\concerto\\concerto\\vtfiles\\PreBiologFile_Rep1_M2.csv", 
                     encoding='latin-1')
ext = pd.read_csv("C:\\Users\\lint730\\concerto\\concerto\\vtfiles\\BioCycID_Processed_Metabolomic_Data_NoRhodOutlier_6_14_24.tsv", delimiter="\t", encoding = 'latin-1')
#We transform the data of the columns where the name of the compounds are on each file to a list.
list_biolog = biolog["Compound"].tolist()
list_ext = ext["Metabolites"].tolist()

#We create a copy of the biolog list.
copy_list_biolog = list_biolog.copy()
#We create an empty dictionary.
compound_dictionary = {}
```

Clean Biolog and metabolomics data for ease of comparison
- dictionary = set of values and a keyword associated with each; good way to map different names for the same thing

```python
#We iterate through each compound on the biolog file.
for i in range(len(list_biolog)):
    #Here we remove special characters and single letters.
    list_biolog[i] = list_biolog[i].lower()
    list_biolog[i] = re.sub(r"\_\-", "", list_biolog[i])
    list_biolog[i] = re.sub(r"\n", "-", list_biolog[i])
    list_biolog[i] = re.sub(r"acid", "-", list_biolog[i])
    list_biolog[i] = re.sub(r" ", "-", list_biolog[i])
    list_biolog[i] = list_biolog[i].split("-")
    list_biolog[i] = [item for item in list_biolog[i] if len(item) > 1]
    #After cleaning, we assign these name to a dictionary.
    compound_dictionary[copy_list_biolog[i]] = list_biolog[i]

#For the exchange metabolomics file we just need to remove the hilpos and hilneg names from the file.
for i in range(len(list_ext)):
    list_ext[i] = list_ext[i].lower()
    list_ext[i] = re.sub(r"\_hilneg", "", list_ext[i])
    list_ext[i] = re.sub(r"\_hilpos", "", list_ext[i])
    #print(list_ext[i])

#We make a copy of the exchange metabolomics file.
copy_list_ext = list_ext.copy()    
```

Select for specific key words and search for presence in exchanged metabolomics list. Produces a list with the key word, and the values that
match in each list
- '''Since the names of the compounds could not be exactly identical, it is necessary to check each match and determine 
if they are the same compound.''' --> automate later


```python
# Here we will compare each keyword with the names of the exchange metabolomics list of names.
Info = "" #empty string variable

for key, values in compound_dictionary.items():
    for keyword in values:
        for compound in list_ext:
            if re.search(keyword, compound):
                #print(keyword)
                Info = Info + key + "," + compound + "\n"
#creat a .csv file, replace name_of_file
new_file = open("name_of_file.csv", "w")
new_file.write(Info)
#if the csv file opened is blank, try to run it again until data shows up. program may be slow to respond/terminates when attempting to open csv
```
